```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('1950-2019//bollywood_full.csv')
```

```python
print(movies.shape)
movies.head(1)
# print(movies.duplicated().sum())
# print(movies.duplicated(subset=['imdb_id','original_title'],keep=False).sum())
# movies[movies.duplicated(subset=['imdb_id','original_title'],keep=False)]
movies = movies.drop_duplicates(subset=['imdb_id','original_title'])
movies.shape
```

```python
# movies = movies.merge(crew,on='imdb_id')
# movies.shape
movies.info()
```

```python
# genres, imdb_id, original_title, story, summary, actors
movies=movies[['imdb_id', 'original_title','story', 'summary','genres', 'actors']]
```

```python
movies.isnull().sum()
```

```python
movies.fillna(value={'story':"",'actors':""},inplace=True)
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
def convert_actors(obj):
    L = []
    counter = 0
    for i in obj.split('|'):
        if counter!=3:
            L.append(i)
            counter+=1
        else:
            break
    return L
```

```python
movies['story'] = movies['story'].apply(lambda x: x.split(' '))
movies['summary'] = movies['summary'].apply(lambda x: x.split(' '))
movies['actors'] = movies['actors'].apply(convert_actors)
movies['genres'] = movies['genres'].apply(lambda x: x.split('|'))
```

```python
movies['actors'] = movies['actors'].apply(lambda x: [i.replace(" ","") for i in x])
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['story'] + movies['summary'] + movies['genres'] + movies['actors']
```

```python
movies.head()
```

```python
df = pd.DataFrame()
df[['movie_id','title','tags']] = movies[['imdb_id','original_title', 'tags']]
```

```python
df.head()
```

```python
df['tags'] = df['tags'].apply(lambda x: " ".join(x))
df['tags'] = df['tags'].apply(lambda x: x.lower())
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
df['tags'] = df['tags'].apply(stem)
```

```python
df['tags'][0]
df.to_csv('file.csv')
```
